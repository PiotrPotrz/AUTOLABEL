```python
from data import DataLoader
from predict_with_dino import PredictWithDino
from visualize import Visualize
from preview import Preview
from tkinter import filedialog
```

<h1>WYBIERZ FOLDER</h1>

```python
# wybierz folder
file_path = filedialog.askdirectory(title="Select a Folder")
```

```python
# załaduj folder
loader = DataLoader(file_path)
```

<h1>DOSTOSUJ PARAMETRY</h1>
<h2>DOSTĘPNE SĄ TRZY PARAMETRY</h2>

1. **TEXT** wpisywany w ramkę, **POWINIEN BYĆ ZAKOŃCZONY KROPKĄ** np. stairs .
2. **BOX THRESHOLD** - najważniejszy parametr - prawdopodobieństwo poprawności predykcji. Im wyższy tym teoretycznie lepsze predykcje. W rzeczywistości nie będzie wykrywać dużej
   części obiektów. Najlepiej w okolicach 0.3 - 0.5 w zależności od foldera.
3. **TEXT THRESHOLD** - opcjonalny, zwykle nie trzeba zmieniać.


<h3>INSTRUKCJA</h3>

1. Wpisz nazwę obiektu.
2. Kliknij SAVE (po każdej zmianie nazwy obiektu)
3. Wybierz thresholdy
4. Zobacz predykcję klikając preview
5. Możesz wybrać inny losowy obraz do podglądu za pomocą ANOTHER IMAGE

**WYCHODZISZ KLIKAJĄĆ KRZYŻYK**. Wszystkie parametry zostaną zapisane

```python
preview = Preview(loader)
```

```python
box_threshold, text_threshold, text = preview.previewer()
```

<h1>LABELOWANIE</h1>

**UWAGA MOŻE ZAJĄĆ DUŻO CZASU**

```python
dino = PredictWithDino()

# LABELOWANIE
dframe = dino.predict_and_save(loader, box_threshold,text_threshold, text)
# ZAPISANIE WYNIKÓW
dframe.to_csv(f"{loader.dir_name}.csv", index=False)
```
